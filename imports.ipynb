```python
# to add the env to the kernal  ---- next time for evalution only need to run this block 
import sys
import os
import subprocess

venv_path = "./cbct_env"
if os.name == "nt":
    python_path = os.path.join(venv_path, "Scripts", "python.exe")
else:
    python_path = os.path.join(venv_path, "bin", "python")

subprocess.run([
    python_path, "-m", "ipykernel", "install",
    "--user",
    "--name=cbct_env",
    "--display-name=Python (cbct_env)"
])
```
